```python
Spender_df=\
    purchase_df.groupby('SN').agg(
        p_cnt=('Purchase ID', 'count'),
        avg_p_price=('Price', 'mean'),
        tot_p_value=('Price', 'sum'))

Spender_df = Spender_df.sort_values('tot_p_value', ascending=False)

Spender_df
```

```python
Popular_df=\
    purchase_df.groupby(['Item ID', 'Item Name']).agg(
        p_cnt=('Purchase ID', 'count'),
        avg_p_price=('Price', 'mean'),
        tot_p_value=('Price', 'sum'))

Popular_df = Popular_df.sort_values('p_cnt', ascending=False)

Popular_df
```
